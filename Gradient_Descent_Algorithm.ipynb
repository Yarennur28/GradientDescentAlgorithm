```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
#from sklearn.preprocessing import MinMaxScaler

#scaler = MinMaxScaler()

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
ds = pd.read_csv("/kaggle/input/house-price-prediction-dataset/House Price Prediction Dataset.csv")
```

```python
X = ds[['Area', 'Bedrooms']].values
Y = ds[['Price']].values

X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)
X_normalized = (X - X_min) / (X_max - X_min)

Y_min = np.min(Y)
Y_max = np.max(Y)
Y_normalized = (Y - Y_min) / (Y_max - Y_min)
```

```python
def hypothesis(X, q):
    return q[0] + q[1] * X[:, 0] + q[2] * X[:, 1]
```

```python
def cost_function(X, Y, q):
    m = len(Y) 
    return (1 / (2 * m)) * np.sum((hypothesis(X, q) - Y.flatten()) ** 2)


def derivative_q0(X, Y, q):
    m = len(Y)
    return (1 / m) * np.sum(hypothesis(X, q) - Y.flatten())

def derivative_q1(X, Y, q):
    m = len(Y)
    return (1 / m) * np.sum((hypothesis(X, q) - Y.flatten()) * X[:, 0])

def derivative_q2(X, Y, q):
    m = len(Y)
    return (1 / m) * np.sum((hypothesis(X, q) - Y.flatten()) * X[:, 1])
```

```python
q = np.array([0.0, 0.0, 0.0])
alpha = 2# Öğrenme oranı
iterations = 100

def gradient_descent(X, Y, alpha, q, iterations):
    
    cost_history = []
    parameters_history = []
    
    for i in range(iterations):
        
        temp0 = q[0] - alpha*derivative_q0(X, Y, q)
        temp1 = q[1] - alpha*derivative_q1(X, Y, q)
        temp2 = q[2] - alpha*derivative_q2(X, Y, q)
        
        q[0] = temp0
        q[1] = temp1
        q[2] = temp2
        
        parameters_history.append(q.copy())
        cost_history.append(cost_function(X, Y, q))
        
        print(f"İterasyon {i + 1}: q0 = {q[0]:.4f}, q1 = {q[1]:.4f}, q2 = {q[2]:.4f}, Maliyet = {cost_history[-1]:.4f}")
        
    return q, cost_history, parameters_history   
        
q_optimal, cost_history, parameters_history = gradient_descent(X_normalized, Y_normalized, alpha, q, iterations)
    
print("Optimal parametreler:", q_optimal)
print("Son maliyet:", cost_history[-1])
```

```python
plt.plot(range(iterations), cost_history, 'b-')
plt.xlabel('İterasyon')
plt.ylabel('Maliyet (Cost)')
plt.title('Maliyet Fonksiyonu Değişimi')
plt.grid()  # Izgara çizgileri ekle
plt.show()
```
